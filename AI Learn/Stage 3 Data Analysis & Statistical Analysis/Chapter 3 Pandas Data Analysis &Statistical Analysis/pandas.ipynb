```python
movie = pd.read_csv('data/movie.csv', index_col='movie_title')
index = movie.index
columns = movie.columns

index_list = index.tolist()     # 索引列表
column_list = columns.tolist()  # 列名列表

index_list[0] = 'Ratava'
index_list[2] = 'Ertceps'
column_list[1] = 'Director Name'
column_list[2] = 'Critical Reviews'
movie.index = index_list
movie.columns = column_list
movie.head()
```

```python
# 3. 添加,删除,插入列
movie = pd.read_csv('data/movie.csv')
movie
```

```python
# 添加列
movie['has_seen'] = 0
movie['all_facebook_likes'] = movie['actor_1_facebook_likes'] + movie['actor_2_facebook_likes'] + movie['actor_3_facebook_likes'] + movie['director_facebook_likes']
movie.head()
```

```python
# 删除列
# inplace 参数, 默认为False, 不修改原对象
movie.drop(labels='all_facebook_likes', axis='columns', inplace=True)
```

```python
movie
```

```python
# 插入列
movie.insert(loc=0,column='profit',value=movie['gross'] - movie['budget'])
movie
```
