```python
# 2. 常用排序方法
movie = pd.read_csv('data/movie.csv')
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.head()
```

```python
# 取前100个imdb_score最大的电影
movie2.nlargest(100, 'imdb_score')
```

```python
# 取前5个budget最小且评分前100的电影
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

```python
# 找到每年imdb评分最高的电影
movie2 = movie[['movie_title', 'title_year', 'imdb_score']]
movie_top_year = movie2.sort_values(['title_year', 'imdb_score'], ascending=False)
final = movie_top_year.drop_duplicates(subset='title_year') # 对年份去重
final.head()
```
